```python
import pandas as pd
import requests
ad_df_url = "https://storage.googleapis.com/karmicseed-assessments-data/ads_performance.json"
response = requests.get(ad_df_url)
data_json =  response.json()
print(data_json[:2])
```

```python
import json
print(json.dumps(data_json[:1], indent=4))
```

```python
ad_df = pd.DataFrame(data_json)
ad_df.head()
```

```python
ad_df.columns
```

```python
ad_df.isna().sum()
```

```python
import numpy as np
MISSING_VALUES = [
    "", " ", "  ", "NA", "N/A", "na", "null", "None", "-", "--"
]
ad_df.replace(MISSING_VALUES, np.nan, inplace=True)
```

```python
ads_numeric_cols = [
    "impressions", "clicks", "costPerClick", "clickThroughRate",
    "cost", "spend",
    "purchases1d", "purchases7d", "purchases14d", "purchases30d",
    "unitsSoldClicks1d", "unitsSoldClicks7d",
    "unitsSoldClicks14d", "unitsSoldClicks30d",
    "sales1d", "sales7d", "sales14d", "sales30d",
    "acosClicks7d", "acosClicks14d",
    "roasClicks7d", "roasClicks14d"
]

ad_df[ads_numeric_cols] = ad_df[ads_numeric_cols].apply(
    pd.to_numeric, errors="coerce"
)
```

```python
ad_df.isna().sum()
```

```python
filtered_rows = ad_df[
    ad_df["costPerClick"].isna() & (ad_df["clicks"] > 0)
]

filtered_rows
```

##### The costPerClick column contains missing values (3708) for rows where the number of clicks is zero. Since cost per click is undefined when no clicks occur, these values represent periods of no ad engagement rather than data quality issues. Therefore, such missing values are safely filled with 0.

```python
ad_df["costPerClick"] = ad_df["costPerClick"].fillna(0)
```

```python
ad_df["clickThroughRate"] = ad_df["clickThroughRate"].fillna(0)
```

```python
ad_df["roasClicks7d"] = ad_df["roasClicks7d"].fillna(0)
ad_df["roasClicks14d"] = ad_df["roasClicks14d"].fillna(0)
```

```python
filtered_rows_acos = ad_df[
    ad_df["acosClicks7d"].isna() ]

filtered_rows_acos
```

```python
acos_agg = (
    ad_df
    .groupby("SKU", as_index=False)
    .agg({
        "acosClicks7d": "mean",
        "acosClicks14d": "mean"
    })
)
acos_agg
```

```python
ad_df = ad_df.merge(
    acos_agg,
    on="SKU",
    how="left",
    suffixes=("", "_sku_avg")
)
```

```python
ad_df["acosClicks14d_clean"] = ad_df["acosClicks14d"].fillna(
    ad_df["acosClicks14d_sku_avg"]
)

ad_df["acosClicks7d_clean"] = ad_df["acosClicks7d"].fillna(
    ad_df["acosClicks7d_sku_avg"]
)
```

```python
ad_df
```

```python
ad_df.drop(["acosClicks7d","acosClicks14d"],axis = "columns",inplace = True)
```

```python
ad_df.isna().sum()
```

```python
ads_agg_df = (
    ad_df
    .groupby("SKU", as_index=False)
    .agg({
        "impressions": "sum",
        "clicks": "sum",
        "cost": "sum",
        "spend": "sum",
        "sales14d": "sum",
        "sales30d": "sum",
        "acosClicks14d_clean": "mean",
        "roasClicks14d": "mean"
    })
    .round(2)
)
```

```python
ads_agg_df
```

```python
ads_agg_df["click_rate"] = np.where(
    ads_agg_df["impressions"] > 0,
    ads_agg_df["clicks"] / ads_agg_df["impressions"],
    0
)
```

```python
ads_agg_df["cost_per_ad_sale"] = np.where(
    ads_agg_df["sales14d"] > 0,
    ads_agg_df["cost"] / ads_agg_df["sales14d"],
    np.nan
)
```

```python
def ads_efficiency(row):
    if row["roasClicks14d"] >= 4:
        return "Efficient Ads"
    elif 2 <= row["roasClicks14d"] < 4:
        return "Moderate Ads"
    elif row["roasClicks14d"] < 2 and row["sales14d"] > 0:
        return "Inefficient Ads"
    else:
        return "No Ad Sales"

ads_agg_df["ads_efficiency"] = ads_agg_df.apply(
    ads_efficiency, axis="columns"
)
```

```python
ads_agg_df["ad_dependency_flag"] = np.where(
    (ads_agg_df["sales14d"] > 0) &
    (ads_agg_df["roasClicks14d"] < 2.5),
    "High Ad Dependency",
    "Normal"
)
```

```python
def pricing_strategy_ads(row):
    if row["ads_efficiency"] == "Efficient Ads":
        return "Allow Price Increase"

    if row["ads_efficiency"] == "Moderate Ads":
        return "Hold Price"

    if row["ads_efficiency"] == "Inefficient Ads":
        return "Reduce Price or Improve Listing"

    if row["ad_dependency_flag"] == "High Ad Dependency":
        return "Freeze Price Increase"

    return "Hold Price"

ads_agg_df["pricing_strategy_ads"] = ads_agg_df.apply(
    pricing_strategy_ads, axis="columns"
)
```

# Ads Performance Processing & Insights

#### The advertising performance data was aggregated at the SKU level to align with pricing decision granularity. Impressions, clicks, ad spend, and ad-attributed sales (14-day and 30-day windows) were summed to capture total exposure, engagement, and revenue contribution from advertising, while efficiency metrics such as 14-day ACOS and ROAS were averaged to reflect typical ad effectiveness rather than short-term spikes. 
#### From this aggregated dataset, several derived metrics were created to assess demand quality and pricing risk. 
#### A click rate metric was calculated as the ratio of clicks to impressions to measure how effectively ad visibility translated into customer interest. 
#### Cost per ad sale was computed by dividing total ad spend by 14-day ad-attributed sales, providing a direct view of how much advertising cost is required to generate one unit of ad-driven revenue. 
#### SKUs were then classified into ad efficiency categories based on 14-day ROAS thresholds, distinguishing products with efficient, moderate, inefficient, or no ad-supported demand. This classification enables clear separation between SKUs where ads confirm strong value perception and those where sales are heavily subsidized by spend. 
#### An ad dependency flag was further introduced to identify SKUs that rely on advertising to generate sales while delivering weak returns, signaling elevated pricing risk.
#### Finally, a pricing strategy recommendation was assigned based on ad efficiency and dependency, explicitly defining whether price increases should be allowed, held, reduced, or frozen. This ensures that pricing decisions are guided by the sustainability of demand rather than inflated by paid traffic.

# Business Summary
#### Advertising performance was used as a demand-quality filter within the pricing framework to ensure price increases are supported by efficient, sustainable demand. SKUs with strong ROAS and low dependency on ad spend were identified as candidates for price increases, while SKUs with inefficient or ad-dependent demand were restricted from upward pricing to prevent margin erosion. By translating raw ad metrics into clear efficiency and risk signals, this approach prevents paid traffic from masking pricing weaknesses and ensures that pricing decisions are grounded in long-term profitability rather than short-term volume.

```python
import pandas as pd
import requests
return_df_url = "https://storage.googleapis.com/karmicseed-assessments-data/returns_data.json"
response = requests.get(return_df_url)
data_json =  response.json()
print(data_json[:2])
```

```python
import json
print(json.dumps(data_json[:2], indent=4))
```

```python
return_data = pd.DataFrame(data_json)
return_data
```

```python
Historical_Conversion_Data = pd.read_csv("Historical_Conversion_Data.csv")
Historical_Conversion_Data.head()
```

```python
returns_df = return_data.copy()
historical_sales_df = Historical_Conversion_Data[["SKU", "Units Ordered","Conversion Rate","Demand Strength"]].copy()
returns_df = returns_df.merge(
    historical_sales_df,
    on="SKU",
    how="left"
)
returns_df
```

```python
returns_df.columns = (
    returns_df.columns
    .str.replace("\n", " ", regex=False)
    .str.strip()
)
returns_df
```

```python
returns_df.isna().sum()
```

```python
import numpy as np
MISSING_VALUES = [
    "", " ", "  ", "NA", "N/A", "na", "null", "None", "-", "--"
]
returns_df.replace(MISSING_VALUES, np.nan, inplace=True)
returns_df
```

```python
returns_df.isna().sum()
```

```python
returns_df["Return Quantity  (Last 7 days)"] = (
    returns_df["Return Quantity  (Last 7 days)"]
    .fillna(0)
)
```

```python
numeric_cols = [
    "Return Quantity  (Last 7 days)",
    "Return Quantity  (Last 30 days)",
    "Return Quantity  (Last 60 days)",
    "Return Quantity  (Last 90 days)"
]
```

```python
for col in numeric_cols:
    returns_df[col] = pd.to_numeric(returns_df[col], errors="coerce")

returns_df.info()
```

```python
returns_df["return_rate_90d"] = (
    returns_df["Return Quantity  (Last 90 days)"]
    / returns_df["Units Ordered"]
)
returns_df
```

```python
# Thresholds (can be tuned later)
HIGH_RETURN_THRESHOLD = 0.08   # 8%
MEDIUM_RETURN_THRESHOLD = 0.03 # 3%
```

```python
def return_risk(rate):
    
    if rate <= MEDIUM_RETURN_THRESHOLD:
        return "Low Return Risk"
    if rate <= HIGH_RETURN_THRESHOLD:
        return "Medium Return Risk"
    return "High Return Risk"

returns_df["return_risk"] = (
    returns_df["return_rate_90d"]
    .apply(return_risk)
)
```

```python
def demand_return_regime(row):

    # Case 1: False pricing signal
    if (
        row["Demand Strength"] == "Strong Demand"
        and row["return_risk"] == "High Return Risk"
    ):
        return "False Demand (High Returns)"

    # Case 2: True pricing power
    if (
        row["Demand Strength"] == "Strong Demand"
        and row["return_risk"] == "Low Return Risk"
    ):
        return "True Demand (Pricing Power)"

    # Case 3: Weak demand but customers are satisfied
    if (
        row["Demand Strength"] == "Weak Demand"
        and row["return_risk"] == "Low Return Risk"
    ):
        return "Low Demand but Good Quality"

    # Case 4: Weak demand and high returns (worst case)
    if (
        row["Demand Strength"] == "Weak Demand"
        and row["return_risk"] == "High Return Risk"
    ):
        return "Weak Demand & High Returns"

    # Case 5: Everything else
    return "Mixed / Watch"
```

```python
returns_df["demand_return_regime"] = returns_df.apply(demand_return_regime, axis="columns")
returns_df
```

```python
returns_df["return_risk"].value_counts()
```

```python
final_returns_signal_df = returns_df[
    [
        "SKU",
        "Units Ordered",
        "Demand Strength",
        "Conversion Rate",
        "return_rate_90d",
        "return_risk",
        "demand_return_regime",
    ]
].round(4)
```

# Business Report
#### This analysis integrates historical conversion performance with return behavior to create a holistic view of true product demand and customer satisfaction at the SKU level. 
#### By combining demand signals (conversion rate and demand strength) with normalized return risk (90-day return rate), the framework distinguishes between apparent demand and sustainable demand. 
#### SKUs with Strong Demand and Low Return Risk are classified as True Demand (Pricing Power), indicating healthy customer acceptance and the ability to support price optimization or margin expansion. 
#### Conversely, Strong Demand with High Return Risk is flagged as False Demand, highlighting cases where high conversion may be driven by aggressive pricing, promotions, or misleading expectations, resulting in elevated returns. 
#### SKUs with Weak Demand but Low Return Risk represent quality-sound products that may benefit from improved visibility, positioning, or marketing rather than price cuts, while Weak Demand with High Return Risk identifies the highest-risk products requiring immediate intervention such as quality fixes, listing corrections, or potential delisting. Overall, this combined demand-return regime enables more precise pricing, merchandising, and product quality decisions by separating true pricing power from demand distortion caused by returns.

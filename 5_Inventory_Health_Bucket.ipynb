```python
import numpy as np
MISSING_VALUES = [
    "", " ", "  ", "NA", "N/A", "na", "null", "None", "-", "--"
]
inventory_df.replace(MISSING_VALUES, np.nan, inplace=True)
inventory_df.isna().sum()
```

```python
inventory_numeric_cols = [
    "total-inventory", "available", "inbound-shipped","inbound-received","reserved-quantity", "unfulfillable-quantity",
    "sell-through", "days-of-supply", "weeks-of-cover-t30", "weeks-of-cover-t90"
    
]

inventory_df[inventory_numeric_cols] = inventory_df[inventory_numeric_cols].apply(
    pd.to_numeric, errors="coerce"
)
```

```python
def inventory_status(row):
    if row["days-of-supply"] < 20:
        return "Low Inventory"
    elif row["days-of-supply"] > 90:
        return "Overstocked"
    else:
        return "Healthy"

inventory_df["inventory_status"] = inventory_df.apply(
    inventory_status, axis="columns"
)
```

```python
inventory_df[["units-shipped-t30", "units-shipped-t90"]] = (
    inventory_df[["units-shipped-t30", "units-shipped-t90"]]
    .apply(pd.to_numeric, errors="coerce")
)
inventory_df["velocity_30d"] = inventory_df["units-shipped-t30"] / 30
inventory_df["velocity_90d"] = inventory_df["units-shipped-t90"] / 90

inventory_df["velocity_trend"] = np.where(
    inventory_df["velocity_30d"] < inventory_df["velocity_90d"],
    "Demand Slowing",
    "Stable / Growing"
)
```

```python
# Mark a SKU as a clearance candidate if it has too much inventory AND demand is slowing.
inventory_df["clearance_candidate"] = np.where(
    (inventory_df["inventory_status"] == "Overstocked") &
    (inventory_df["velocity_trend"] == "Demand Slowing"),
    "Yes",
    "No"
)
```

```python
def pricing_strategy_inventory(row):
    # Case 1: Strong demand + healthy inventory --> Increase price
    if (
        row["inventory_status"] == "Healthy"
        and row["sell-through"] > 1.2
        and row["days-of-supply"] < 80
    ):
        return "Increase Price"

    # Case 2: Very strong demand but inventory slightly high --> Small increase
    if (
        row["inventory_status"] == "Healthy"
        and row["sell-through"] > 1.4
        and row["days-of-supply"] >= 80
    ):
        return "Small Increase Price"

    # Case 3: Balanced SKU → Hold price
    if (
        row["inventory_status"] == "Healthy"
        and 0.9 <= row["sell-through"] <= 1.2
    ):
        return "Hold Price"

    # Case 4: Overstocked but demand strong --> Let inventory absorb naturally
    if (
        row["inventory_status"] == "Overstocked"
        and row["sell-through"] > 1.3
        and row["velocity_trend"] == "Stable / Growing"
    ):
        return "Hold Price (Absorb Inventory)"

    # Case 5: Clearance candidate → Reduce price
    if row["clearance_candidate"] == "Yes":
        return "Reduce Price (Clearance)"

    # Case 6: Operational issue (inventory exists but not available)
    if (
        row["inventory_status"] == "Overstocked"
        and row["available"] == 0
    ):
        return "Freeze Pricing,Fix Availability"

    # Default
    return "Hold Price"

inventory_df["pricing_strategy_inventory"] = inventory_df.apply(
    pricing_strategy_inventory,
    axis="columns"
)
```

#### Inventory health analysis was used as a critical control layer to ensure pricing decisions balance revenue optimization with operational sustainability. By examining inventory coverage (days of supply), sales velocity trends, sell-through performance, and availability constraints, we identified how supply conditions should influence both the direction and urgency of pricing actions. 
#### SKUs with low "days of supply" were recognized as being at risk of stock-outs, requiring pricing protection to moderate demand and preserve service levels. 
#### Conversely, SKUs with excess inventory and extended coverage periods were flagged for demand stimulation, as prolonged holding increases storage costs, working-capital lock-up, and aging risk. 
#### Importantly, inventory levels were interpreted in conjunction with sales velocity trends rather than in isolation. Overstocked SKUs with stable or growing demand were allowed to absorb inventory naturally without unnecessary price reductions, preserving margin where demand momentum remained healthy. 
#### In contrast, SKUs exhibiting both high inventory coverage and slowing sales velocity were proactively identified as clearance candidates, enabling early, controlled price reductions to reduce long-term inventory risk rather than relying on reactive liquidation. 
#### Balanced SKUs with healthy inventory and moderate sell-through were assigned a price-hold strategy to maintain stability and avoid introducing unnecessary demand volatility.
#### Additionally, cases where inventory existed but was not available for sale were treated as operational issues rather than pricing problems, highlighting the importance of resolving fulfillment constraints before making price adjustments.
#### Overall, this inventory-led pricing framework ensures that pricing actions are forward-looking, risk-aware, and aligned with supply realities, preventing short-term revenue optimization from creating long-term operational inefficiencies. By embedding inventory health directly into pricing decisions, the business can improve cash-flow efficiency, reduce aging and storage exposure, and apply price increases or reductions only where they are operationally justified and financially sustainable.

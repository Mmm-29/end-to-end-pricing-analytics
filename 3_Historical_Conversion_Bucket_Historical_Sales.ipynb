```python
import requests
url = "https://storage.googleapis.com/karmicseed-assessments-data/historical_sales.json"
response = requests.get(url)
data_json =  response.json()
print(data_json[:2])
```

```python
import json
print(json.dumps(data_json[:2], indent=4))
```

```python
import pandas as pd
Historical_Sales = pd.DataFrame(data_json)
Historical_Sales
```

```python
Historical_Sales.tail()
```

```python
import numpy as np
MISSING_VALUES = [
    "", " ", "  ", "NA", "N/A", "na", "null", "None", "-", "--"
]
Historical_Sales.replace(MISSING_VALUES, np.nan, inplace=True)
Historical_Sales
```

```python
Historical_Sales.isna().sum()
```

```python
Historical_Sales.info()
```

```python
Historical_Sales.columns
```

```python
numeric_cols = [
    "Sessions - Total",
    "Page Views - Total",
    "Units Ordered",
    "Ordered Product Sales",
    "Total Order Items"
]
```

```python
for col in numeric_cols:
    Historical_Sales[col] = (
        Historical_Sales[col]
        .astype(str)
        .str.replace(",", "", regex=False)
        .str.strip()
    )
```

```python
for col in numeric_cols:
    Historical_Sales[col] = pd.to_numeric(Historical_Sales[col], errors="coerce")
```

```python
Historical_Sales["Date"] = pd.to_datetime(Historical_Sales["Date"], errors="coerce")
```

```python
Historical_Sales.info()
```

```python
Historical_Sales.head()
```

```python

Historical_Sales[numeric_cols] = Historical_Sales[numeric_cols].fillna(0)
Historical_Sales.isna().sum()
```

```python
# Aggregate Demand Metrics Per SKU
df = (
    Historical_Sales
    .groupby("SKU", as_index=False)
    .agg({
        "Sessions - Total": "sum",
        "Units Ordered": "sum",
        "Ordered Product Sales": "sum",
        "Page Views - Total": "sum",
        "Total Order Items" : "sum"
    })
)
df
```

```python

df["Conversion Rate"] = ((df["Units Ordered"] / df["Sessions - Total"]) * 100).round(2)
df
```

```python
conversion_median = df["Conversion Rate"].median()
conversion_median
```

```python
df["Revenue per Session"] =  round(((df["Ordered Product Sales"] / df["Sessions - Total"]) * 100),2)
df
```

```python
df["Demand Strength"] = np.where(
    df["Conversion Rate"] >= conversion_median,
    "Strong Demand",
    "Weak Demand"
)
df
```

```python
df.to_csv("Historical_Conversion_Data.csv",index = False)
```

# Business Report
#### This analysis consolidates historical sales data at the SKU level to evaluate customer demand and sales efficiency across the product portfolio. 
#### By aggregating sessions, page views, units ordered, order items, and total sales, the dataset provides a holistic view of traffic, conversion, and revenue performance for each SKU.
#### Conversion Rate is used as the primary demand signal, reflecting how effectively customer visits translate into purchases. 
#### SKUs are classified into Strong Demand or Weak Demand segments based on whether their conversion rate is above or below the portfolio median, enabling relative performance benchmarking rather than arbitrary thresholds.
#### Additionally, Revenue per Session highlights monetization efficiency by showing how much revenue each customer visit generates. 
#### Together, these metrics help identify high-performing SKUs that combine strong demand and efficient revenue generation, as well as underperforming products where traffic exists but conversion or monetization is weak—supporting data-driven decisions on pricing, promotions, inventory planning, and product-level optimization.

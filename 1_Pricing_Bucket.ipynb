```python
import pandas as pd
import requests
pricing_df_url ="https://storage.googleapis.com/karmicseed-assessments-data/pricing_data.json" 
response = requests.get(pricing_df_url)
data_json =  response.json()
print(data_json[:2])
```

```python
import json
print(json.dumps(data_json[:2], indent=4))
```

```python
pricing_df = pd.DataFrame(data_json)
pricing_df
```

```python
import numpy as np
MISSING_VALUES = [
    "", " ", "  ", "NA", "N/A", "na", "null", "None", "-", "--"
]
pricing_df.replace(MISSING_VALUES, np.nan, inplace=True)
pricing_df
```

```python
numeric_cols = [
    "FBA Fee",
    "Storage Fee",
    "Handling_Cost",
    "Cost",
    "Current_Price",
    "Minimum_Acceptable_Margin_%",
    "Target_Gross_Margin_%"
]

for col in numeric_cols:
    pricing_df[col] = (
        pricing_df[col]
        .astype(str)
        .str.replace("%", "", regex=False)
        .str.replace(",", "", regex=False)
        .str.replace("", "", regex=False)
        .str.replace("$", "", regex=False)
        .str.strip()
    )
```

```python
for col in numeric_cols:
    pricing_df[col] = pd.to_numeric(pricing_df[col], errors="coerce")
pricing_df
```

```python
pricing_df.isna().sum()
```

```python
pricing_df["Handling_Cost"] = pricing_df["Handling_Cost"].fillna(0.75)
```

```python
pricing_df.isna().sum()
```

```python
pricing_df["unit_cost"] = (
    pricing_df["Cost"]
    + pricing_df["FBA Fee"]
    + pricing_df["Storage Fee"]
    + pricing_df["Handling_Cost"]
)

```

```python
# Normalize Margin Percentages
for col in ["Minimum_Acceptable_Margin_%", "Target_Gross_Margin_%"]:
    pricing_df[col] = pricing_df[col] / 100
```

```python
#Min Price = Unit Cost / (1 - Minimum Margin)

pricing_df["min_price"] = (
    pricing_df["unit_cost"] / (1 - pricing_df["Minimum_Acceptable_Margin_%"])
)
```

```python
pricing_df["target_price"] = (
    pricing_df["unit_cost"] / (1 - pricing_df["Target_Gross_Margin_%"])
)
```

```python
pricing_df["loss_amount_per_unit"] = (pricing_df["unit_cost"] - pricing_df["Current_Price"]).clip(lower=0)
pricing_df["loss_making"] = (pricing_df["unit_cost"] - pricing_df["Current_Price"]) > 0
```

```python
# Margin violation flag
pricing_df["margin_violation"] = (
    pricing_df["Current_Price"] < pricing_df["min_price"]
)
```

```python
pricing_df
```

```python
final_pricing_df = pricing_df[
    [
        "SKU",
        "Product_description",
        "Product Role",
        "unit_cost",
        "min_price",
        "target_price",
        "Current_Price",
        "margin_violation",
        "loss_amount_per_unit",
        "loss_making",
    ]
]
```

```python
final_pricing_df.to_csv("cost_margin_bucket.csv", index=False)
```

## Business Insights
#### This pricing analysis evaluates the profitability and margin compliance of the current SKU portfolio by calculating the total landed unit cost (including product cost, fulfillment, storage, and handling fees) and comparing it against current selling prices. For each SKU, minimum acceptable and target prices were derived based on defined gross margin thresholds, enabling clear identification of margin violations. The analysis distinguishes between margin erosion and true loss-making products, ensuring that profitable but under-margined SKUs are not incorrectly classified as losses. The resulting dataset provides a focused, decision-ready view of pricing health across core and seasonal products, supporting data-driven actions such as targeted price corrections, margin optimization, and prioritization of SKUs requiring immediate intervention.
